```python
def DistanceKm(lat_or, lon_or, lat_des, lon_des):
    lat_or = float(lat_or)
    lon_or = float(lon_or)
    lat_des = float(lat_des)
    lon_des = float(lon_des)

    rad_earth = 6371 # radius of the earth
    
    lat_or = lat_or * np.pi/180
    lat_des = lat_des * np.pi/180
    lon_or = lon_or * np.pi/180
    lon_des = lon_des * np.pi/180
    
    lat = lat_des - lat_or
    lon = lon_des - lon_or
    
    a = np.sin(lat/2)**2 + np.cos(lat_or) * np.cos(lat_des) * np.sin(lon/2)**2
    c = 2 * np.arctan2(np.sqrt(a),np.sqrt(1-a))

    # a = sin²(Δlat/2) + cos(lat1) · cos(lat2) · sin²(Δlong/2)
    # c = 2 · atan2(√a, √(1−a))
    distance = c * rad_earth
    return distance
```

```python
max_distance = 0
min_distance = 10000
key = ""
closest = ["",""]
far = ["",""]
dic = {}
for i in range(len(dataset['bts_id'])):
  distances = []
  for j in range(len(dataset['bts_id'])):
    distance = DistanceKm(dataset['lat'][i],dataset['lon'][i],dataset['lat'][j],dataset['lon'][j])
    #distances.append(distance)
    if distance > max_distance:
      far = [dataset['lat'][j],dataset['lon'][j]]
      max_distance = distance
    if distance < min_distance:
      closest = [dataset['lat'][j],dataset['lon'][j]]
      min_distance = distance
  
  #dic = {str((dataset['lat'][i], dataset['lon'][i])): far, closest}
  key = str((dataset['lat'][i] , dataset['lon'][i]))
  dic[key] = far, closest
```
