```python
import pandas as pd
```

```python
df_hu = pd.read_csv('bts_comuna.csv')
df_carlos = pd.read_csv('antennas_geolocation.csv')
```

```python
df_hu.info()
```

```python
df_hu = df_hu.sort_values(by='bts_id')
#df_hu = df_hu.reset_index()

df_carlos = df_carlos.sort_values(by='bts_id')
#df_carlos = df_carlos.reset_index()
```

```python
if(set(df_carlos['bts_id'].unique().tolist()) == set(df_hu['bts_id'].unique().tolist())):
    print('Set is Corret')
else:
    print('Set is Wrong')
```

```python
import unidecode

count_comuna_error = 0

for count in range(df_carlos['bts_id'].count()):
    if(df_hu['bts_id'][count] != df_carlos['bts_id'][count]):
        print('index error')
        break

    if(abs(df_hu['lat'][count] - df_carlos['lat'][count]) > 0.0001):
        print('lat error ' + df_carlos['bts_id'][count])
        #break

    if(abs(df_hu['lon'][count] - df_carlos['lon'][count]) > 0.001):
        print('lon error')
        #break
    
    if(not pd.isna(df_carlos['comuna'][count])):
        if(unidecode.unidecode(df_hu['comuna'][count]).upper() != unidecode.unidecode(df_carlos['comuna'][count]).upper()):
            print('comuna error ' + df_carlos['bts_id'][count] + ' ' + df_hu['comuna'][count] + ' vs ' + df_carlos['comuna'][count])
            count_comuna_error += 1
            #break
```

```python
print('comuna error num ' + str(count_comuna_error))
```
